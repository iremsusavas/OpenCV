# Pedestrian Detection using Haar Cascade Classifiers

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def show(image = None, title = "Image"):
    size = 16
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
video_capture = cv2.VideoCapture("videos/pedestrians.mp4")

body_cascade = cv2.CascadeClassifier('Haar/haarcascade_fullbody.xml')



while True:  # Loop once video is successfully loaded
    ret,frame = video_capture.read()
    gray_ = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    pedestrians = body_cascade.detectMultiScale(gray_,1.2,3)
    for (x,y,w,h) in pedestrians: # Extract bounding boxes for any bodies identified
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,0),2)
        cv2.putText(frame,'Pedestrian',(x,y),cv2.FONT_HERSHEY_PLAIN,1,(255,255,0),2)
    out.write(frame)
    cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):   #PLEASE PRESS "q" TO QUIT FROM THE CAMERA
        break
        
        
cv2.destroyAllWindows()
video_capture.release()
show(frame,"Pedestrians")
```
